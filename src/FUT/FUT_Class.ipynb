```python
# @title
import pandas as pd
import numpy as np


# Função para adicionar colunas de pesos
def add_weights(data, sw):
    n = len(data)
    if sw == 'Linear':
        # Pesos lineares
        data['weight_linear'] = np.linspace(1, 2, n)
    elif sw == 'Log(x) + x':
        # Pesos baseados em log(x) + x normalizado no intervalo [1, 2]
        linspace = np.linspace(1, n, n)  # Intervalo de 1 a N
        weights = np.log(linspace) + linspace
        weights = (weights - weights.min()) / (weights.max() - weights.min())  # Normalização
        weights = weights * (2 - 1) + 1  # Escalando para [1, 2]
        data['weight_log_x_plus_x'] = weights
    return data

def gen_all():
  # Carregar e concatenar os datasets por liga
  df_base_EPL = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/EPL/football_results.csv')
  df_24_EPL = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/EPL/football_results_2024.csv')
  df_EPL = pd.concat([df_base_EPL, df_24_EPL], ignore_index=True)

  df_base_LaLiga = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/LaLiga/football_results.csv')
  df_24_LaLiga = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/LaLiga/football_results_2024.csv')
  df_LaLiga = pd.concat([df_base_LaLiga, df_24_LaLiga], ignore_index=True)

  df_base_Bundesliga = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Bundesliga/football_results.csv')
  df_24_Bundesliga = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Bundesliga/football_results_2024.csv')
  df_Bundesliga = pd.concat([df_base_Bundesliga, df_24_Bundesliga], ignore_index=True)

  df_base_SerieA = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/SerieA/football_results.csv')
  df_24_SerieA = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/SerieA/football_results_2024.csv')
  df_SerieA = pd.concat([df_base_SerieA, df_24_SerieA], ignore_index=True)

  df_base_Ligue1 = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Ligue1/football_results.csv')
  df_24_Ligue1 = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Ligue1/football_results_2024.csv')
  df_Ligue1 = pd.concat([df_base_Ligue1, df_24_Ligue1], ignore_index=True)

  # Adicionar colunas de pesos a cada dataset
  df_EPL = add_weights(df_EPL, 'Linear')
  df_EPL = add_weights(df_EPL, 'Log(x) + x')

  df_LaLiga = add_weights(df_LaLiga, 'Linear')
  df_LaLiga = add_weights(df_LaLiga, 'Log(x) + x')

  df_Bundesliga = add_weights(df_Bundesliga, 'Linear')
  df_Bundesliga = add_weights(df_Bundesliga, 'Log(x) + x')

  df_SerieA = add_weights(df_SerieA, 'Linear')
  df_SerieA = add_weights(df_SerieA, 'Log(x) + x')

  df_Ligue1 = add_weights(df_Ligue1, 'Linear')
  df_Ligue1 = add_weights(df_Ligue1, 'Log(x) + x')

  # Concatenar todos os datasets em um único dataframe
  df_all = pd.concat([df_EPL, df_LaLiga, df_Bundesliga, df_SerieA, df_Ligue1], ignore_index=True)

  # Converter a coluna "date" para datetime
  df_all['date'] = pd.to_datetime(df_all['date'])

  # Ordenar o dataframe pelo campo "date"
  df_all = df_all.sort_values(by='date').reset_index(drop=True)

  # # Salvar o resultado
  df_all.to_csv('/content/drive/MyDrive/Datasets/FUT/all_leagues.csv', index=False)

  # Carregar e concatenar os datasets por liga
  df_base_EPL_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/EPL/football_results_.csv')
  df_24_EPL_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/EPL/football_results_2024_.csv')
  df_EPL_ = pd.concat([df_base_EPL_, df_24_EPL_], ignore_index=True)

  df_base_LaLiga_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/LaLiga/football_results_.csv')
  df_24_LaLiga_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/LaLiga/football_results_2024_.csv')
  df_LaLiga_ = pd.concat([df_base_LaLiga_, df_24_LaLiga_], ignore_index=True)

  df_base_Bundesliga_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Bundesliga/football_results_.csv')
  df_24_Bundesliga_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Bundesliga/football_results_2024_.csv')
  df_Bundesliga_ = pd.concat([df_base_Bundesliga_, df_24_Bundesliga_], ignore_index=True)

  df_base_SerieA_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/SerieA/football_results_.csv')
  df_24_SerieA_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/SerieA/football_results_2024_.csv')
  df_SerieA_ = pd.concat([df_base_SerieA_, df_24_SerieA_], ignore_index=True)

  df_base_Ligue1_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Ligue1/football_results_.csv')
  df_24_Ligue1_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Ligue1/football_results_2024_.csv')
  df_Ligue1_ = pd.concat([df_base_Ligue1_, df_24_Ligue1_], ignore_index=True)

  # Adicionar colunas de pesos a cada dataset
  df_EPL_ = add_weights(df_EPL_, 'Linear')
  df_EPL_ = add_weights(df_EPL_, 'Log(x) + x')

  df_LaLiga_ = add_weights(df_LaLiga_, 'Linear')
  df_LaLiga_ = add_weights(df_LaLiga_, 'Log(x) + x')

  df_Bundesliga_ = add_weights(df_Bundesliga_, 'Linear')
  df_Bundesliga_ = add_weights(df_Bundesliga_, 'Log(x) + x')

  df_SerieA_ = add_weights(df_SerieA_, 'Linear')
  df_SerieA_ = add_weights(df_SerieA_, 'Log(x) + x')

  df_Ligue1_ = add_weights(df_Ligue1_, 'Linear')
  df_Ligue1_ = add_weights(df_Ligue1_, 'Log(x) + x')

  # Concatenar todos os datasets em um único dataframe
  df_all_ = pd.concat([df_EPL_, df_LaLiga_, df_Bundesliga_, df_SerieA_, df_Ligue1_], ignore_index=True)

  # Converter a coluna "date" para datetime
  df_all_['date'] = pd.to_datetime(df_all_['date'])

  # Ordenar o dataframe pelo campo "date"
  df_all_ = df_all_.sort_values(by='date').reset_index(drop=True)

  # # Salvar o resultado
  df_all_.to_csv('/content/drive/MyDrive/Datasets/FUT/all_leagues_.csv', index=False)

gen_all()

# Carregar e concatenar os datasets por liga
df_pred_EPL = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/EPL/football_pred.csv')
df_pred_LaLiga = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/LaLiga/football_pred.csv')
df_pred_Bundesliga = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Bundesliga/football_pred.csv')
df_pred_SerieA = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/SerieA/football_pred.csv')
df_pred_Ligue1 = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Ligue1/football_pred.csv')

# Concatenar todos os datasets em um único dataframe
df_pred_all = pd.concat([df_pred_EPL, df_pred_LaLiga, df_pred_Bundesliga, df_pred_SerieA, df_pred_Ligue1], ignore_index=True)

# Converter a coluna "date" para datetime
df_pred_all['date'] = pd.to_datetime(df_pred_all['date'])

# Ordenar o dataframe pelo campo "date"
df_pred_all = df_pred_all.sort_values(by='date').reset_index(drop=True)

# Salvar o resultado
df_pred_all.to_csv('/content/drive/MyDrive/Datasets/FUT/all_pred.csv', index=False)

# Carregar e concatenar os datasets por liga
df_pred_EPL_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/EPL/football_pred_.csv')
df_pred_LaLiga_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/LaLiga/football_pred_.csv')
df_pred_Bundesliga_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Bundesliga/football_pred_.csv')
df_pred_SerieA_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/SerieA/football_pred_.csv')
df_pred_Ligue1_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/Ligue1/football_pred_.csv')

# Concatenar todos os datasets em um único dataframe
df_pred_all_ = pd.concat([df_pred_EPL_, df_pred_LaLiga_, df_pred_Bundesliga_, df_pred_SerieA_, df_pred_Ligue1_], ignore_index=True)

# Converter a coluna "date" para datetime
df_pred_all_['date'] = pd.to_datetime(df_pred_all_['date'])

# Ordenar o dataframe pelo campo "date"
df_pred_all_ = df_pred_all_.sort_values(by='date').reset_index(drop=True)

# Salvar o resultado
df_pred_all_.to_csv('/content/drive/MyDrive/Datasets/FUT/all_pred_.csv', index=False)
```

```python
# @title
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from scipy.stats import mode
import numpy as np
from IPython.display import display_html

def create_result_btts(df):
    # Criando a coluna `result`
    df['result'] = df.apply(
        lambda row: 1 if row['home_goals'] > row['away_goals'] else
                    2 if row['home_goals'] < row['away_goals'] else 0,
        axis=1
    )

    # Criando a coluna `both_teams_score`
    df['both_teams_score'] = df.apply(
        lambda row: 1 if row['home_goals'] > 0 and row['away_goals'] > 0 else 0,
        axis=1
    )

    return df

def train_models_rf(data, sample_weight):
    # Separar features e alvos
    X = data.drop(columns=['result', 'both_teams_score'])
    y_result = data['result']
    y_both_teams_score = data['both_teams_score']

    # Treinar os modelos
    model_result = RandomForestClassifier(n_estimators=100, random_state=42)
    model_result.fit(X, y_result, sample_weight=sample_weight)

    model_both_teams_score = RandomForestClassifier(n_estimators=100, random_state=42)
    model_both_teams_score.fit(X, y_both_teams_score, sample_weight=sample_weight)

    return model_result, model_both_teams_score

def train_models_xgb(data, sample_weight):
    # Separar features e alvos
    X = data.drop(columns=['result', 'both_teams_score'])
    y_result = data['result']
    y_both_teams_score = data['both_teams_score']

    # Treinar os modelos
    # model_result = XGBClassifier(objective='multi:softmax', random_state=42)
    model_result = GradientBoostingClassifier(n_estimators=100, random_state=42)
    model_result.fit(X, y_result, sample_weight=sample_weight)

    # model_both_teams_score = XGBClassifier(objective='binary:logistic', random_state=42)
    model_both_teams_score = GradientBoostingClassifier(n_estimators=100, random_state=42)
    model_both_teams_score.fit(X, y_both_teams_score, sample_weight=sample_weight)

    return model_result, model_both_teams_score

def preprocess_new_data(new_data):
    features_to_drop = ['home_team_name', 'away_team_name', 'match_id', 'date', 'home_goals', 'away_goals']
    return new_data.drop(columns=features_to_drop)

def make_predictions(new_data_path, model_result, model_both_teams_score):
    new_data = pd.read_csv(new_data_path)

    if len(new_data) > 0:
        processed_data = preprocess_new_data(new_data)

        # Fazer predições
        predictions_result = model_result.predict(processed_data)
        predictions_both_teams_score = model_both_teams_score.predict(processed_data)

        # Adicionar predições ao DataFrame
        new_data['predicted_result'] = predictions_result
        new_data['predicted_both_teams_score'] = predictions_both_teams_score

        return new_data
    else:
        return pd.DataFrame(columns=['date', 'home_team_name', 'away_team_name', 'predicted_result', 'predicted_both_teams_score'])

# Consolidando resultados
consolidated_results = []
df_all = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/all_leagues.csv')
df_all_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/all_leagues_.csv')

df_all = create_result_btts(df_all)
df_all_ = create_result_btts(df_all_)

for data_f in ['', '_']:
    for sw in ['Linear', 'Log(x) + x']:
        data = df_all if data_f == '' else df_all_

        # Remover colunas irrelevantes
        data = data.drop(columns=['home_team_name', 'away_team_name', 'match_id', 'date', 'home_goals', 'away_goals'])

        if sw == 'Linear':
            sample_weight = data['weight_linear']
            data = data.drop(columns=['weight_linear', 'weight_log_x_plus_x'])
            model_result_rf, model_both_rf = train_models_rf(data, sample_weight)
            model_result_xgb, model_both_xgb = train_models_xgb(data, sample_weight)

        elif sw == 'Log(x) + x':
            sample_weight = data['weight_log_x_plus_x']
            data = data.drop(columns=['weight_linear', 'weight_log_x_plus_x'])
            model_result_rf, model_both_rf = train_models_rf(data, sample_weight)
            model_result_xgb, model_both_xgb = train_models_xgb(data, sample_weight)

        # Fazer previsões para Random Forest e XGBoost
        result_rf = make_predictions(f'/content/drive/MyDrive/Datasets/FUT/all_pred{data_f}.csv', model_result_rf, model_both_rf)
        result_xgb = make_predictions(f'/content/drive/MyDrive/Datasets/FUT/all_pred{data_f}.csv', model_result_xgb, model_both_xgb)

        # Adicionar resultados
        consolidated_results.append(result_rf[['predicted_result', 'predicted_both_teams_score']])
        consolidated_results.append(result_xgb[['predicted_result', 'predicted_both_teams_score']])

# Calcular a moda e a confiança final dos resultados acumulados
all_results = pd.concat(consolidated_results, axis=1)

# DataFrame para armazenar os resultados finais
final_result = pd.DataFrame()

# Moda para "predicted_result"
result_modes = mode(all_results.iloc[:, ::2], axis=1)
final_result['predicted_result'] = result_modes.mode.flatten()

# Confiança para "predicted_result"
result_counts = result_modes.count.flatten()  # Frequência da moda
total_models_result = all_results.iloc[:, ::2].shape[1]  # Total de modelos
final_result['confidence_result'] = (result_counts / total_models_result) * 100

# Moda para "predicted_both_teams_score"
both_modes = mode(all_results.iloc[:, 1::2], axis=1)
final_result['predicted_both_teams_score'] = both_modes.mode.flatten()

# Confiança para "predicted_both_teams_score"
both_counts = both_modes.count.flatten()  # Frequência da moda
total_models_both = all_results.iloc[:, 1::2].shape[1]  # Total de modelos
final_result['confidence_both_teams_score'] = (both_counts / total_models_both) * 100

# Adicionar informações do DataFrame original
final_result = pd.concat([result_rf[['date', 'home_team_name', 'away_team_name']], final_result], axis=1)

final_result.to_csv('/content/drive/MyDrive/Datasets/FUT/result1.csv')

# Exibir resultados
display_html(
    final_result.to_html(index=False),
    raw=True
)
```

```python
# @title
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from scipy.stats import mode
import numpy as np
from IPython.display import display_html

path_dict = {
    'EPL': 'EPL',
    'La Liga': 'LaLiga',
    'Bundesliga': 'Bundesliga',
    'Serie A': 'SerieA',
    'Ligue 1': 'Ligue1',
    }

leagues = ['EPL', 'La Liga', 'Bundesliga', 'Serie A', 'Ligue 1']

def create_result_btts(df):
    # Criando a coluna `result`
    df['result'] = df.apply(
        lambda row: 1 if row['home_goals'] > row['away_goals'] else
                    2 if row['home_goals'] < row['away_goals'] else 0,
        axis=1
    )

    # Criando a coluna `both_teams_score`
    df['both_teams_score'] = df.apply(
        lambda row: 1 if row['home_goals'] > 0 and row['away_goals'] > 0 else 0,
        axis=1
    )

    return df

def train_models_rf(data, sample_weight):
    # Separar features e alvos
    X = data.drop(columns=['result', 'both_teams_score'])
    y_result = data['result']
    y_both_teams_score = data['both_teams_score']

    # Treinar os modelos
    model_result = RandomForestClassifier(n_estimators=100, random_state=42)
    model_result.fit(X, y_result, sample_weight=sample_weight)

    model_both_teams_score = RandomForestClassifier(n_estimators=100, random_state=42)
    model_both_teams_score.fit(X, y_both_teams_score, sample_weight=sample_weight)

    return model_result, model_both_teams_score

def train_models_xgb(data, sample_weight):
    # Separar features e alvos
    X = data.drop(columns=['result', 'both_teams_score'])
    y_result = data['result']
    y_both_teams_score = data['both_teams_score']

    # Treinar os modelos
    # model_result = XGBClassifier(objective='multi:softmax', random_state=42)
    model_result = GradientBoostingClassifier(n_estimators=100, random_state=42)
    model_result.fit(X, y_result, sample_weight=sample_weight)

    # model_both_teams_score = XGBClassifier(objective='binary:logistic', random_state=42)
    model_both_teams_score = GradientBoostingClassifier(n_estimators=100, random_state=42)
    model_both_teams_score.fit(X, y_both_teams_score, sample_weight=sample_weight)

    return model_result, model_both_teams_score

def preprocess_new_data(new_data):
    features_to_drop = ['home_team_name', 'away_team_name', 'match_id', 'date', 'home_goals', 'away_goals']
    return new_data.drop(columns=features_to_drop)

def make_predictions(new_data_path, model_result, model_both_teams_score):
    new_data = pd.read_csv(new_data_path)

    if len(new_data) > 0:
        processed_data = preprocess_new_data(new_data)

        # Fazer predições
        predictions_result = model_result.predict(processed_data)
        predictions_both_teams_score = model_both_teams_score.predict(processed_data)

        # Adicionar predições ao DataFrame
        new_data['predicted_result'] = predictions_result
        new_data['predicted_both_teams_score'] = predictions_both_teams_score

        return new_data
    else:
        return pd.DataFrame(columns=['date', 'home_team_name', 'away_team_name', 'predicted_result', 'predicted_both_teams_score'])

g = pd.DataFrame()

for league in leagues:
  # Consolidando resultados
  consolidated_results = []
  # df_all = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/all_leagues.csv')
  # df_all_ = pd.read_csv('/content/drive/MyDrive/Datasets/FUT/all_leagues_.csv')

  # df_all = create_result_btts(df_all)
  # df_all_ = create_result_btts(df_all_)

  for data_f in ['', '_']:
      for sw in ['Linear', 'Log(x) + x']:
          df_pre = pd.read_csv(f'/content/drive/MyDrive/Datasets/FUT/{path_dict[league]}/football_results{data_f}.csv')
          df_24 = pd.read_csv(f'/content/drive/MyDrive/Datasets/FUT/{path_dict[league]}/football_results_2024{data_f}.csv')
          data = pd.concat([df_pre, df_24], ignore_index=True)
          data = create_result_btts(data)
          # data = df_all if data_f == '' else df_all_

          # Remover colunas irrelevantes
          data = data.drop(columns=['home_team_name', 'away_team_name', 'match_id', 'date', 'home_goals', 'away_goals'])

          if sw == 'Linear':
              sample_weight = np.linspace(1, 2, len(data))  # Pesos de 1 (antigo) a 2 (recente)
              # data = data.drop(columns=['weight_linear', 'weight_log_x_plus_x'])
              model_result_rf, model_both_rf = train_models_rf(data, sample_weight)
              model_result_xgb, model_both_xgb = train_models_xgb(data, sample_weight)

          elif sw == 'Log(x) + x':
              # Função ajustada para calcular os pesos com base em (log(x) + x) normalizada no intervalo [1, 2]
              sample_weights_ = np.linspace(1, 2, len(data))
              sample_weights_ = np.log(sample_weights_) + sample_weights_

              # Normalizando os pesos para o intervalo [1, 2]
              sample_weights_ = 1 + (sample_weights_ - sample_weights_.min()) / (sample_weights_.max() - sample_weights_.min())
              sample_weight = sample_weights_
              # data = data.drop(columns=['weight_linear', 'weight_log_x_plus_x'])
              model_result_rf, model_both_rf = train_models_rf(data, sample_weight)
              model_result_xgb, model_both_xgb = train_models_xgb(data, sample_weight)

          # Fazer previsões para Random Forest e XGBoost
          result_rf = make_predictions(f'/content/drive/MyDrive/Datasets/FUT/{path_dict[league]}/football_pred{data_f}.csv', model_result_rf, model_both_rf)
          result_xgb = make_predictions(f'/content/drive/MyDrive/Datasets/FUT/{path_dict[league]}/football_pred{data_f}.csv', model_result_xgb, model_both_xgb)

          # Adicionar resultados
          consolidated_results.append(result_rf[['predicted_result', 'predicted_both_teams_score']])
          consolidated_results.append(result_xgb[['predicted_result', 'predicted_both_teams_score']])

  # Calcular a moda e a confiança final dos resultados acumulados
  all_results = pd.concat(consolidated_results, axis=1)

  # DataFrame para armazenar os resultados finais
  final_result = pd.DataFrame()

  # Moda para "predicted_result"
  result_modes = mode(all_results.iloc[:, ::2], axis=1)
  final_result['predicted_result'] = result_modes.mode.flatten()

  # Confiança para "predicted_result"
  result_counts = result_modes.count.flatten()  # Frequência da moda
  total_models_result = all_results.iloc[:, ::2].shape[1]  # Total de modelos
  final_result['confidence_result'] = (result_counts / total_models_result) * 100

  # Moda para "predicted_both_teams_score"
  both_modes = mode(all_results.iloc[:, 1::2], axis=1)
  final_result['predicted_both_teams_score'] = both_modes.mode.flatten()

  # Confiança para "predicted_both_teams_score"
  both_counts = both_modes.count.flatten()  # Frequência da moda
  total_models_both = all_results.iloc[:, 1::2].shape[1]  # Total de modelos
  final_result['confidence_both_teams_score'] = (both_counts / total_models_both) * 100

  # Adicionar informações do DataFrame original
  final_result = pd.concat([result_rf[['date', 'home_team_name', 'away_team_name']], final_result], axis=1)
  g = pd.concat([g, final_result], axis=0)

  # Exibir resultados
  display_html(
      f'''
      <div style="display: flex; justify-content: center;">
          <h3>League: {league}</h3>
      </div>
      ''',
      raw=True
  )
  display_html(
      final_result.to_html(index=False),
      raw=True
  )
g.to_csv('/content/drive/MyDrive/Datasets/FUT/result2.csv')
```
