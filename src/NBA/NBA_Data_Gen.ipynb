```python
base_url = "https://www.nba.com/news/power-rankings-2024-25-week-"
for i in range(1,26):
  # print(i)
  r = requests.get(base_url + str(i))
  html_doc = r.text
  soup = BeautifulSoup(html_doc, 'html.parser')

  for team in soup.find_all("div", {"class": "ArticlePowerRankings_pr__payHO"}):
    try:
      Name = team.find("a", {"class": "Anchor_anchor__cSc3P ArticlePowerRankings_prTeam__dTyqP Link_styled__okbXW"}).getText()
      for p in team.find("div", {"class": "ArticleContent_article__NBhQ8"}).find_all("p"):
        if "Pace:" in p.getText():
          Pace = p.getText().split("Pace:")[1].split("(")[0].strip()
          OffRtg = p.getText().split("OffRtg:")[1].split("(")[0].strip()
          DefRtg = p.getText().split("DefRtg:")[1].split("(")[0].strip()
          NetRtg = p.getText().split("NetRtg:")[1].split("(")[0].strip()
      GOW = team.find("div", {"class": "ArticleContent_article__NBhQ8"}).find_all("p")[-1].getText() if any(v in team.find("div", {"class": "ArticleContent_article__NBhQ8"}).find_all("p")[-1].getText() for v in ["vs.", "@"]) else "-"
      # new_row = {"Season":"23-24", "Week": i, "Name": Name, "Pace": Pace, "OffRtg": OffRtg, "DefRtg": DefRtg, "NetRtg": NetRtg, "GOW": GOW}
      new_row = pd.DataFrame([{"Season":"23-24", "Week": i, "Name": Name, "Pace": Pace, "OffRtg": OffRtg, "DefRtg": DefRtg, "NetRtg": NetRtg, "GOW": GOW}])
      print(new_row)
      # df_24_25 = df_24_25.append(new_row, ignore_index=True)
      df_24_25 = pd.concat([df_24_25, new_row], ignore_index=True)
    except Exception as e:
      print("Week:", i, "Name:", "All:", team.find("div", {"class": "ArticleContent_article__NBhQ8"}).find("p").getText())
      # print(f"ERROR >>> {e}")
      break
```

```python
# Carrega o JSON da URL
url = 'https://fixturedownload.com/feed/json/nba-2024'
response = requests.get(url)
data = response.json()

# Converte o JSON para DataFrame
nba_sc_24 = pd.DataFrame(data)

# Renomeia as colunas para se assemelhar ao CSV original
nba_sc_24 = nba_sc_24.rename(columns={
    'MatchNumber': 'Match Number',
    'RoundNumber': 'Round Number',
    'DateUtc': 'Date',
    'Location': 'Location',
    'HomeTeam': 'Home Team',
    'AwayTeam': 'Away Team',
    'Result': 'Result'
})

# Ajusta a coluna Date para o formato desejado considerando UTC-3
nba_sc_24['Date'] = pd.to_datetime(nba_sc_24['Date']).apply(
    lambda x: (x - timedelta(hours=3)).strftime('%d/%m/%Y %H:%M')
)

# Calcula a coluna Result evitando erros com NaN
nba_sc_24['Result'] = nba_sc_24.apply(
    lambda x: f"{int(x['HomeTeamScore'])} - {int(x['AwayTeamScore'])}"
    if pd.notna(x['HomeTeamScore']) and pd.notna(x['AwayTeamScore']) else pd.NA,
    axis=1
)

# Remove colunas desnecessárias
nba_sc_24 = nba_sc_24.drop(columns=['Group', 'HomeTeamScore', 'AwayTeamScore'], errors='ignore')

# Reorganiza as colunas se necessário
colunas_ordenadas = ['Match Number', 'Round Number', 'Date', 'Location', 'Home Team', 'Away Team', 'Result']
nba_sc_24 = nba_sc_24[colunas_ordenadas]

nba_specific = df_specific
```
